```python
class Model(object):
    def __init__(self, first_name, last_name=None, age=None):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
    
    def get_full_name(self):
        if self.last_name:
            return (
                f'''
                {self.first_name} {self.last_name}
                '''
            )
        return (
            f'''
            {self.first_name}
            '''
        )

    def __repr__(self):
        return self.get_full_name()
        


model = Model('someone')
print(model)
```
