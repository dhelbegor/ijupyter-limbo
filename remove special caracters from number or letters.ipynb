```python
# https://www.dotnetperls.com/in-python

from string import punctuation

def transform(s):
    "A simple but efficient string transform to remove special characters."
    return ''.join(c for c in s if c not in punctuation)

n = transform('055.824.771-23')
l = transform("this is a simple string test?, (yes it is), don't be afraid, test it. we are the boss!")

print(n)
print(l)

for x in punctuation:
    print('[%s]' % (x))
```
