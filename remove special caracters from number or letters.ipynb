```python
# https://www.dotnetperls.com/in-python
def transform(s):
    from string import punctuation
    "A simple but efficient string transform to remove special characters."
    return ''.join(c for c in s if c not in punctuation)

n = transform('055.824.771-23')
l = transform("this is a simple string test?, (yes it is), don't be afraid, test it. we are the boss!")

print(n)
print(l)


#for x in punctuation:
#    print('[%s]' % (x))
```

```python
# A simple method allowed by django to remove special characters from a number.
#https://docs.python.org/3/library/functools.html

def clean_phone(phone): # in django needs to have self to.
    import re
    #from functools import reduce
    #return reduce(lambda x, y: x+y, re.findall(r'\d', phone))
    
    return re.sub('[^0-9]', '', phone)
    
phones = [
    '18878482967', '+18878482967', '+1 887 8482967', '+1 (887) 8482-967'
]

for phone in phones:
    print(clean_phone(phone))
```
