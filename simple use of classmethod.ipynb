```python
from datetime import date

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    @classmethod
    def from_birth_year(cls, name, birth_year):
        print("from: %s" %(cls.__name__.lower()))
        return cls(name, date.today().year - birth_year)
    
    def display(self):
        print(self.name + "'s age is: " + str(self.age))
        
person = Person.from_birth_year('dhelbegor', 1949)
person.display()

person1 = Person.from_birth_year('Fernanda', 1990)
person1.display()
```
