```python
from xml.dom import minidom
from xml.etree import ElementTree as xml

import pandas as pd




doc = xml.parse('items.xml')
tree = doc.getroot()

def transform(key, item):
    return key.find(item).text

def to_line(line, itens):
    return [transform(line, item) for item in itens]


item = [
    'author',
    'title',
    'genre',
    'price',
    'publish_date'
]

df = pd.DataFrame([to_line(linha, item) for linha in tree], columns=item)

display(df)


# items = doc.getElementsByTagName('author')

# for elem in items:
#     # print(elem.attributes['name'].value)
#     print(elem.firstChild.data)


```
